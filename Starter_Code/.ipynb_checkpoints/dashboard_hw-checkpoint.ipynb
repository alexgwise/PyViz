# San Francisco Rental Prices Dashboard

In this notebook, you will compile the visualizations from the previous analysis into functions that can be used for a Panel dashboard.

```python
# imports
import panel as pn
pn.extension('plotly')
import plotly.express as px
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import os
from pathlib import Path
from dotenv import load_dotenv
```

```python
# Read the Mapbox API key
load_dotenv()
mapbox_token = os.getenv("MAPBOX_API_KEY")
len(mapbox_token)
# Set the Mapbox API
px.set_mapbox_access_token(mapbox_token)
```

# Import Data

```python
# Import the CSVs to Pandas DataFrames
file_path = Path("Data/sfo_neighborhoods_census_data.csv")
sfo_data = pd.read_csv(file_path, index_col="year")

file_path = Path("Data/neighborhoods_coordinates.csv")
df_neighborhood_locations = pd.read_csv(file_path)
```

## Panel Visualizations

In this section, you will copy the code for each plot type from your analysis notebook and place it into separate functions that Panel can use to create panes for the dashboard. 

These functions will convert the plot object to a Panel pane.

Be sure to include any DataFrame transformation/manipulation code required along with the plotting code.

Return a Panel pane object from each function that can be used to build the dashboard.

Note: Remove any `.show()` lines from the code. We want to return the plots instead of showing them. The Panel dashboard will then display the plots.

```python
# Calculate the data - Housing Units per Year
housing_units = sfo_data[["housing_units"]]
housing_avg = housing_units.groupby("year").mean()
housing_avg = housing_avg.rename(columns={"housing_units":"Average Housing Units"})
housing_avg = housing_avg.reset_index()
```

```python
# Calculate the data - Gross Rent per Year
gross_rent = sfo_data[['gross_rent']]
avg_gross_rent = gross_rent.groupby("year").mean()
```

```python
# Calculate the data - Average Sales Price per Year
sales_price = sfo_data[['sale_price_sqr_foot']]
avg_sales_price = sales_price.groupby('year').mean()
avg_sales_price = avg_sales_price.rename(columns={"sale_price_sqr_foot":"Sales Price per Square Foot"})
avg_sales_price = avg_sales_price.reset_index()
```

```python
# Calculate the data - Average Prices by Neighborhood
avg_price_by_neigh = sfo_data.reset_index()
avg_price_by_neigh = avg_price_by_neigh.rename(columns={"sale_price_sqr_foot":"Sales Price per Square Foot"})
```

```python
# Calculate the data - Top 10 Most Expensive Neighborhoods
top_10 = sfo_data.groupby('neighborhood').mean()
top_10 = top_10.sort_values(by="sale_price_sqr_foot", ascending=False)
top_10 = top_10.iloc[0:10]
top_10 = top_10.reset_index()
#rename columns
top_10 = top_10.rename(columns={"sale_price_sqr_foot":"Sales Price per Square Foot"})
```

```python
# Calculate the data - Neighborhood coordinates
neighborhood_mean_values = sfo_data.groupby('neighborhood').mean()
neighborhood_mean_values = neighborhood_mean_values.reset_index()
neighborhood_concat = pd.concat([df_neighborhood_locations,neighborhood_mean_values],axis="columns",join="inner")
neighborhood_concat = neighborhood_concat.drop(columns="neighborhood")
```

```python
# Visualize Housing Units Data
housing_units_plot = housing_avg.plot(
    kind="bar", 
    title="Housing Units in San Francisco from 2010 to 2016",
    x="year",
    y="Average Housing Units",
    figsize=(10,7),
    ylim=(370000,387500))
```

```python
# Visualize Gross Rent Data
avg_gross_rent_plot = avg_gross_rent.plot(
    title="Average Gross Rent in San Francisco", figsize=(10,7))
```

```python
# Visualize Average Sales Price
sales_price_plot = avg_sales_price.plot(
    title="Average Sales Price per Square Foot in San Francisco",
    x="year",
    y="Sales Price per Square Foot",
    figsize=(10,7))
```

```python
# Visualize Average Price by Neighborhood
avg_neigh_price_plot = avg_price_by_neigh.hvplot(
    x='year',
    y='Sales Price per Square Foot',
    groupby='neighborhood')
```

```python
# Visualize Top 10 Most Expensive Neighborhoods
top_10_plot = top_10.hvplot.bar(
    x="neighborhood", 
    y="Sales Price per Square Foot", 
    title = "Top 10 Most Expensive Neighborhoods in San Fran",
    rot=45, 
    height=500,
    ylim=(600,950))
```

```python
# Visualize Parallel Coordinates Plot
parallel_coor = px.parallel_coordinates(top_10, color="Sales Price per Square Foot")
```

```python
# Visualize Parallel Categories Plot
parallel_cat = px.parallel_categories(top_10, color="Sales Price per Square Foot")
```

```python
# Visualize Neighborhood Map
neighborhood_map = px.scatter_mapbox(
    neighborhood_concat,
    lat="Lat",
    lon="Lon",
    size="sale_price_sqr_foot",
    color="gross_rent",
    title="Average Sale Price per Square Foot and Gross Rent for San Fran Neighborhoods")
```

## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
#containerize each tab
welcome_tab = pn.Column(
    "#Welcome to the San Fran Real Estate Analysis Dashboard!",
    "##2010 - 2016",
    "This dashboard provides a visual analysis of historical prices of house units, sales price per square foot, and gross rent.\n You can use the tabs above to navigate through the various visualizations.",
    neighborhood_map)

yearly_analysis_tab = pn.Column(
    housing_units_plot.figure,
    avg_gross_rent_plot.figure,
    sales_price_plot.figure)


neigh_analysis_tab = pn.Column(
    avg_neigh_price_plot,
    top_10_plot)

parallel_tab = pn.Column(
    parallel_coor,
    parallel_cat)

panel_tabs = pn.Tabs(
    ("Welcome", welcome_tab),
    ("Yearly Market Analysis", yearly_analysis_tab),
    ("Neighborhood Analysis", neigh_analysis_tab),
    ("Parallel Plots", parallel_tab))
    
```

## Serve the Panel Dashboard

```python
panel_tabs.servable()
```
